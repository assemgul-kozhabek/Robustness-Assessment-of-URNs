```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a random graph
G = nx.erdos_renyi_graph(20, 0.2)

# Plot initial graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=500)
plt.title("Initial Graph")
plt.show()

# Calculate local efficiency before any removal
initial_local_efficiency = nx.local_efficiency(G)
print("Initial local efficiency:", round(initial_local_efficiency, 4))

# Remove nodes based on betweenness centrality (highest first) and calculate local efficiency after each removal
while G.number_of_nodes() > 0:
    betweenness = nx.betweenness_centrality(G)
    node_to_remove = max(betweenness, key=betweenness.get)

    # Color the node with the highest betweenness centrality in red
    node_colors = ['red' if node == node_to_remove else 'lightblue' for node in G.nodes]

    # Plot graph after coloring the node and print local efficiency
    plt.figure(figsize=(8, 6))
    nx.draw(G, with_labels=True, node_color=node_colors, node_size=500)
    plt.title(f"Graph after coloring node {node_to_remove} in red")

    G.remove_node(node_to_remove)

    if G.number_of_nodes() > 1:
        local_efficiency = nx.local_efficiency(G)
        plt.text(0, -0.1, f"Local efficiency: {round(local_efficiency, 4)}", transform=plt.gca().transAxes)
        print(f"Removed node {node_to_remove}.")
    else:
        print(f"Removed node {node_to_remove}. Graph has only one node left.")

    plt.show()
```
